```python
import numpy as np
import pandas as pd

# importing visualization tools
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPool2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
## keras dataset
from keras.datasets import fashion_mnist
```

```python
## Loading dataset
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
print(x_train.shape)
print(y_train.shape)
print()
print(x_test.shape)
print(y_test.shape)
```

```python
class_labels = ["T-shirt/top","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]
```

```python
## we can also see the images present in the dataset

plt.figure(figsize=(15,10))
j = 1
for i in np.random.randint(0, 1000,25):
    plt.subplot(5,5,j); j+=1
    plt.imshow(x_test[i], cmap='Reds')
```

```python
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)
```

```python
x_train.ndim
x_train.shape
```

```python
## Now we have to normalize the features

x_train = x_train/255
x_test = x_test/255
```

```python
# spliting the data
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x_train ,y_train , test_size=0.3, random_state=42)
```

```python
x_train.shape
```

```python
## training a data
cnn = Sequential()
cnn.add(Conv2D(filters=32, kernel_size=3, strides=(1,1), padding="valid", activation='relu', input_shape=(28,28,1)))
cnn.add(MaxPool2D(pool_size=(2,2), strides=2))

cnn.add(Conv2D(filters=32, kernel_size=3, strides=(1,1), padding="valid", activation='relu'))
cnn.add(MaxPool2D(pool_size=(2,2), strides=2))


cnn.add(Flatten())

cnn.add(Dense(units=128, activation='relu'))
cnn.add(Dense(units=10,activation='softmax'))
```

```python
cnn.summary()
```

```python
cnn.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics = ["accuracy"])
```

```python
cnn.fit(x_train, y_train, batch_size=32, epochs=25, verbose=True, validation_data=(x_val, y_val))
```

```python
# making predictions
y_train_pred = cnn.predict(x_train)
```

```python
y_train_pred.round(2)
```

```python
## evaluating model
cnn.evaluate(x_train, y_train)
```

```python
## evaluating model
cnn.evaluate(x_test, y_test)
```
