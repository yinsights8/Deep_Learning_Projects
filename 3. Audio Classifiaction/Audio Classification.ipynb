```python
## this is the library which is use to display the audio data
import IPython.display as ipd

# scipy.io is also the library which with the help we can prepross the audio data but it dosen't normalize the data
from scipy.io import wavfile as wav
```

### Audio data preprocessing with librosa library

```python
filepath = "free-spoken-digit-dataset-master/recordings/0_george_1.wav"

plt.figure(figsize=(12,6))

data, sample_rate = librosa.load(filepath)
librosa.display.waveshow(data, sr=sample_rate)
ipd.Audio(filename=filepath)
```

```python
# if we are preprocessing the audio data then the values also get normalized 
print(data)

# and kgh is also increased
print(sample_rate)
```

### Audio data preprocessing with the help of scipy library

```python
wav_sample_rate, wav_data = wav.read(filepath)

# as we can see that the wav_data is not normalized, so we need to normalize the data
print(wav_data)

# this is frequency of sample_rate in kghz
print(wav_sample_rate)
```
